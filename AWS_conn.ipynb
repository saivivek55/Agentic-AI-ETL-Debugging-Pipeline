```python
# Cell 1: Setup
import os
import urllib.request
import psycopg2
from psycopg2 import sql, errors

HOST = "database-1.cjwu80qg2q9m.us-east-2.rds.amazonaws.com"
PORT = 5432
ADMIN_DB = "postgres"
USER = "etl_admin"
PASSWORD = "Agenticproject"   # better: os.getenv("PGPASSWORD")
NEW_DB = "ev_etl_db"

CERT_PATH = os.path.join(os.getcwd(), "global-bundle.pem")
CERT_URL = "https://truststore.pki.rds.amazonaws.com/global/global-bundle.pem"

if not os.path.exists(CERT_PATH):
    urllib.request.urlretrieve(CERT_URL, CERT_PATH)

print("CA bundle ready:", CERT_PATH)
```

```python
# Cell 2: Connect to postgres (admin DB)
conn = psycopg2.connect(
    host=HOST,
    port=PORT,
    dbname=ADMIN_DB,
    user=USER,
    password=PASSWORD,
    sslmode="verify-full",
    sslrootcert=CERT_PATH,
)
conn.autocommit = True
print("Connected to postgres")
```

```python
# Cell 3: Create database if not exists
with conn.cursor() as cur:
    cur.execute("SELECT 1 FROM pg_database WHERE datname = %s;", (NEW_DB,))
    exists = cur.fetchone() is not None

    if exists:
        print(f"Database '{NEW_DB}' already exists.")
    else:
        cur.execute(sql.SQL("CREATE DATABASE {}").format(sql.Identifier(NEW_DB)))
        print(f"Database '{NEW_DB}' created.")
```

```python
# Cell 4: Verify database creation
with conn.cursor() as cur:
    cur.execute("SELECT datname FROM pg_database WHERE datname = %s;", (NEW_DB,))
    row = cur.fetchone()

if row:
    print("Verified: database exists ->", row[0])
else:
    print("Database not found.")
```
